```python
# ======================
# Tutorial 03 - Part 1
# Google Colab and Python
# ======================
import numpy as np
import pandas as pd

# 1. Create a numpy array of size 2 x 2
array_2x2 = np.array([[1, 2], [3, 4]])
print('1Ô∏è‚É£ 2x2 array:\n', array_2x2)

# 2. Create a random numpy array of size 3 x 4
random_3x4 = np.random.rand(3, 4)
print('\n2Ô∏è‚É£ Random 3x4 array:\n', random_3x4)

# 3. Create a numpy array of ones and a numpy array of zeroes
ones_array = np.ones((2, 3))
zeros_array = np.zeros((3, 2))
print('\n3Ô∏è‚É£ Array of ones:\n', ones_array)
print('\nArray of zeroes:\n', zeros_array)

# 4. Create a random numpy array of size 6x3 and transpose it
rand_6x3 = np.random.rand(6, 3)
print('\n4Ô∏è‚É£ Original 6x3 array:\n', rand_6x3)
print('\nTransposed array (3x6):\n', rand_6x3.T)

# 5. Create two random numpy arrays of size n x p and p x m, then compute the dot product
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)
print('\n5Ô∏è‚É£ Dot product of A (2x3) and B (3x4):\n', dot_product)
```
